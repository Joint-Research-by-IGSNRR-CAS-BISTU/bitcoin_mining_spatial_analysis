```python
import libpysal
import numpy as np
from math import log10
np.random.seed(10)
e_w = libpysal.io.open(r'E:\new_projects\bitdata\last2\Spatial weight matrix excluding diagonal.gal').read()
from esda.moran import Moran, Moran_BV
```

```python
cap_all = [log10(i+1) for i in new_sf['Cap.All']]
cap_fos = [log10(i+1) for i in new_sf['Cap.Fos']]
cap_ren = [log10(i+1) for i in new_sf['Cap.Ren']]
```

```python
cap_all
```

```python
mbi_all = Moran_BV(df['Hash_rate_standardized'], cap_all, e_w, transformation='r', permutations=999)
mbi_fos = Moran_BV(df['Hash_rate_standardized'], cap_fos, e_w, transformation='r', permutations=999)
mbi_ren = Moran_BV(df['Hash_rate_standardized'], cap_ren, e_w, transformation='r', permutations=999)
print("bivariate Molan's I of all :", round(mbi_all.I, 5), round(mbi_all.p_sim, 5), round(mbi_all.z_sim, 5))
print("bivariate Molan's I of fos :", round(mbi_fos.I, 5), round(mbi_fos.p_sim, 5), round(mbi_fos.z_sim, 5))
print("bivariate Molan's I of ren :", round(mbi_ren.I, 5), round(mbi_ren.p_sim, 5), round(mbi_ren.z_sim, 5))
```

```python
mbi_all = Moran_BV(df['Hash_rate_standardized'], cap_all, i_w, transformation='r', permutations=999)
mbi_fos = Moran_BV(df['Hash_rate_standardized'], cap_fos, i_w, transformation='r', permutations=999)
mbi_ren = Moran_BV(df['Hash_rate_standardized'], cap_ren, i_w, transformation='r', permutations=999)
print("bivariate Molan's I of all :", round(mbi_all.I, 5), round(mbi_all.p_sim, 5), round(mbi_all.z_sim, 5))
print("bivariate Molan's I of fos :", round(mbi_fos.I, 5), round(mbi_fos.p_sim, 5), round(mbi_fos.z_sim, 5))
print("bivariate Molan's I of ren :", round(mbi_ren.I, 5), round(mbi_ren.p_sim, 5), round(mbi_ren.z_sim, 5))
```
